```python
#Практическая 11

#В последовательности на \(n\) целых чисел умножить все элементы на последний
# минимальный элемент.

n = int(input("Введите количество элементов n: "))

numbers = []
for i in range(n):
    numbers.append(int(input(f"Введите {i+1}-е число: ")))

min_val = min(numbers[::-1])

numbers = [x * min_val for x in numbers]

print("Готовый список:", numbers)
```

```python
#Составить генератор (yield), который переведет символы строки из верхнего
#регистра в нижний.

def to_lower_generator(string):
    for char in string:
        yield char.lower()

user_str = input("Введите строку: ")

result_str = "".join(to_lower_generator(user_str))

print("Результат работы:", result_str)
```
