```python
#Практическая 8

#Отсортируйте словарь {'Physics':82, 'Math':65, 'history':75} по значению
#по возрастанию
dano = {'Physics': 82, 'Math': 65, 'history': 75}

otsortirovano = sorted(dano, key=dano.get)

otvet = {}
otvet[otsortirovano[0]] = dano[otsortirovano[0]]
otvet[otsortirovano[1]] = dano[otsortirovano[1]]
otvet[otsortirovano[2]] = dano[otsortirovano[2]]

print("Исходный словарь:", dano)
print("Отсортированный словарь:", otvet)


#Дан словарь на 6 персон, найти и вывести наибольшее и наименьшее значение
#роста (в см.). (Пример, {"Андрей": 178, "Виктор": 150, "Максим": 200, ...},
#наибольшее 200, наименьшее 150).

rost = {
    "Андрей": 178,
    "Виктор": 150,
    "Максим": 200,
    "Елена": 165,
    "Дмитрий": 185,
    "Анна": 170
}

samiy_bolshoy = max(rost.values())
samiy_malenkiy = min(rost.values())

print("наибольшее", samiy_bolshoy, ", наименьшее", samiy_malenkiy)
```
